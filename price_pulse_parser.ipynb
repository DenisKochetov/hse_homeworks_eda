```python
import requests, os
from bs4 import BeautifulSoup
import pandas as pd
from datetime import date
import ast

# Получение таблицы по запросу
def parse_url(url):
    head = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
    }
    response = requests.get(url, headers=head)
    soup = BeautifulSoup(response.content, 'html.parser')
    mydivs = soup.find_all("div", {"class": "pi-body js-pi-body"})
    # prices = mydivs.find_all("div", {"class": "pi-price"})
    return mydivs

# Поиск таблиц на сайте чепионат по заданию
def pulse(url):
    url = "https://www.pulscen.ru/price/0704-kabeli-i-provoda"
    divs = parse_url(url) 
    return divs
```

```python
# b-pagination-listItem номера страниц
```

```python
all_prods = pulse()
```

```python
def get_prod_table(all_prods):
  output = pd.DataFrame()
  for prod in all_prods:
    ga_dict = ast.literal_eval(prod.find("a", {"class": "pi-title js-bp-title js-ga-link js-catalogue-ecommerce js-conversion-event"})["data-ecommerce-ga"])
    prod_dict = ga_dict['ecommerce']['click']['products'][0]
    prod_dict['currency'] = ga_dict['ecommerce']['currencyCode']
    output = output.append(prod_dict, ignore_index=True)
  return output

```

```python
get_prod_table(all_prods)
```
